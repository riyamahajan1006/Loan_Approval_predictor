```python
import kagglehub
import pandas as pd
import os
path = kagglehub.dataset_download(
    "architsharma01/loan-approval-prediction-dataset"
)
file_path = os.path.join(path, "loan_approval_dataset.csv")
df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df[" education"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df[" education"]=le.fit_transform(df[" education"])
le2=LabelEncoder()
df[" loan_status"]=le2.fit_transform(df[" loan_status"])
le3=LabelEncoder()
df[" self_employed"]=le3.fit_transform(df[" self_employed"])
```

```python
df.head(3)
```

```python
df.drop(columns=["loan_id"], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head(2)
```

```python
y.head(2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)   
x_test=scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
rf=RandomForestClassifier()
sv=SVC()
lr=LogisticRegression()
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
rf.fit(x_train, y_train)
y_pred1=rf.predict(x_test)   
print("Accuracy:", accuracy_score(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
# choosen
```

```python
sv.fit(x_train, y_train)
y_pred2=sv.predict(x_test)  
print("Accuracy:", accuracy_score(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

```python
lr.fit(x_train, y_train)
y_pred3=lr.predict(x_test)  
print("Accuracy:", accuracy_score(y_test, y_pred3))
print(classification_report(y_test, y_pred3))
```

```python
import joblib
joblib.dump(le, "education_encoder.pkl")
joblib.dump(le2, "loan_status_encoder.pkl")
joblib.dump(le3, "self_employed_encoder.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(rf, "model.pkl")
```
